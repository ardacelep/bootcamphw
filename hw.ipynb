```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Housing.csv')
```

```python
sns.scatterplot(x=df["price"],data=df,y=df["area"]) # price-area ilikisi
```

```python
sns.barplot(y=df["price"],data=df,x=df["bedrooms"]) # price-bedroom ilikisi
```

```python
sns.barplot(y=df["price"],data=df,x=df["stories"]) # price-stories ilikisi
```

```python
# evleri bathroom saysna gre grupladk, price deerlerinin ortalamasn aldk ve bathroom saylarn index'ten
# tekrar stuna dntrdk.
bathroom_avg_price = df.groupby('bathrooms')['price'].mean().reset_index()

plt.figure(figsize=(8, 6)) # tablonun boyut ayarlamalar
sns.barplot(x='bathrooms', y='price', data=bathroom_avg_price, palette='viridis') # stun grafii ve deerleri
plt.title('Average House Price by Number of Bathrooms') # grafik bal
plt.xlabel('Number of Bathrooms') # x ekseni etiketi
plt.ylabel('Average Price') # y ekseni etiketi
plt.show() # tabloyu gsterme fonksiyonu
```

```python
# verileri furnishing durumuna gre grupladk, durumlarn price deerlerinin ortalamasn alp index'lerini 
# daha iyi bir grnm iin stuna evirdik.
furnishing_avg_price = df.groupby('furnishingstatus')['price'].mean().reset_index()

plt.figure(figsize=(8, 6)) # grafik boyutlarnn ayarlanmas.
sns.barplot(x='furnishingstatus', y='price', data=furnishing_avg_price, palette='plasma') # stun grafii ve deerleri
plt.title('Average House Price by Furnishing Status') # grafik bal
plt.xlabel('Furnishing Status') # x ekseni etiketi
plt.ylabel('Average Price') # y ekseni etiketi
plt.show() # tabloyu gsterme fonksiyonu
```

```python
# burada stunlardan sadece saysal veri olanlar filtreliyoruz, nk korelasyon matriksleri sadece saysal
# veriler iin anlamldr.
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# korelasyon matrisinin oluturulmas
correlation_matrix = numeric_df.corr()

# matrisin grselletirilmesi
plt.figure(figsize=(10, 8))

# Korelasyon matrisini daha grsel hale getirmek iin Seaborn'daki "heatmap" fonksiyonu kullanlr. 
# Renkler ve annot=True ile her hcredeki korelasyon deerini gsteririz.
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)

plt.title('Correlation Matrix of Housing Data')
plt.show()
```
