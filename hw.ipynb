```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Housing.csv')
```

```python
sns.scatterplot(x=df["price"],data=df,y=df["area"]) # price-area ilişkisi
```

```python
sns.barplot(y=df["price"],data=df,x=df["bedrooms"]) # price-bedroom ilişkisi
```

```python
sns.barplot(y=df["price"],data=df,x=df["stories"]) # price-stories ilişkisi
```

```python
# evleri bathroom sayısına göre grupladık, price değerlerinin ortalamasını aldık ve bathroom sayılarını index'ten
# tekrar sütuna dönüştürdük.
bathroom_avg_price = df.groupby('bathrooms')['price'].mean().reset_index()

plt.figure(figsize=(8, 6)) # tablonun boyut ayarlamaları
sns.barplot(x='bathrooms', y='price', data=bathroom_avg_price, palette='viridis') # sütun grafiği ve değerleri
plt.title('Average House Price by Number of Bathrooms') # grafik başlığı
plt.xlabel('Number of Bathrooms') # x ekseni etiketi
plt.ylabel('Average Price') # y ekseni etiketi
plt.show() # tabloyu gösterme fonksiyonu
```

```python
# verileri furnishing durumuna göre grupladık, durumların price değerlerinin ortalamasını alıp index'lerini 
# daha iyi bir görünüm için sütuna çevirdik.
furnishing_avg_price = df.groupby('furnishingstatus')['price'].mean().reset_index()

plt.figure(figsize=(8, 6)) # grafik boyutlarının ayarlanması.
sns.barplot(x='furnishingstatus', y='price', data=furnishing_avg_price, palette='plasma') # sütun grafiği ve değerleri
plt.title('Average House Price by Furnishing Status') # grafik başlığı
plt.xlabel('Furnishing Status') # x ekseni etiketi
plt.ylabel('Average Price') # y ekseni etiketi
plt.show() # tabloyu gösterme fonksiyonu
```

```python
# burada sütunlardan sadece sayısal veri olanları filtreliyoruz, çünkü korelasyon matriksleri sadece sayısal
# veriler için anlamlıdır.
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# korelasyon matrisinin oluşturulması
correlation_matrix = numeric_df.corr()

# matrisin görselleştirilmesi
plt.figure(figsize=(10, 8))

# Korelasyon matrisini daha görsel hale getirmek için Seaborn'daki "heatmap" fonksiyonu kullanılır. 
# Renkler ve annot=True ile her hücredeki korelasyon değerini gösteririz.
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)

plt.title('Correlation Matrix of Housing Data')
plt.show()
```
